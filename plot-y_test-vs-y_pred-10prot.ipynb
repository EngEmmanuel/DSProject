```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
y_test = pd.read_csv('y_test.csv',delimiter='\t')
ylr_pred = pd.read_csv('ylr_pred.csv',delimiter='\t')
yrfr_pred = pd.read_csv('yrfr_pred.csv',delimiter='\t')
ygb_pred = pd.read_csv('ygb_pred.csv',delimiter='\t')
```

```python
# fig, ax = plt.subplots(1,3)
# ax[0].scatter(y_test,ylr_pred)
# ax[1].scatter(y_test,yrfr_pred)
# ax[2].scatter(y_test,ygb_pred)
```

```python
# Linear Regression
plt.scatter(y_test,ylr_pred, marker='x')

lims = [
    np.min([plt.xlim(), plt.ylim()]),  # min of both axes
    np.max([plt.xlim(), plt.ylim()]),  # max of both axes
]

plt.plot(lims, lims, 'k-')
plt.xlabel('pChEBML Values – Test')
plt.ylabel('pChEBML Values – Predicted')
plt.title('Linear Regression')
plt.tight_layout()
plt.savefig('lin-reg-y-10prot.png')
```

```python
plt.cla()
# Random Forest
plt.scatter(y_test,yrfr_pred, marker='x')

lims = [
    np.min([plt.xlim(), plt.ylim()]),  # min of both axes
    np.max([plt.xlim(), plt.ylim()]),  # max of both axes
]

plt.plot(lims, lims, 'k-')
plt.xlabel('pChEBML Values – Test')
plt.ylabel('pChEBML Values – Predicted')
plt.title('Random Forest')
plt.tight_layout()
plt.savefig('random-forest-y-10prot.png')
```

```python
plt.cla()
# Gradient Boost
plt.scatter(y_test,ygb_pred, marker='x')

lims = [
    np.min([plt.xlim(), plt.ylim()]),  # min of both axes
    np.max([plt.xlim(), plt.ylim()]),  # max of both axes
]

plt.plot(lims, lims, 'k-')
plt.xlabel('pChEBML Values – Test')
plt.ylabel('pChEBML Values – Predicted')
plt.title('Gradient Boost')
plt.tight_layout()
plt.savefig('gboost-y-10prot.png')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
y_test = pd.read_csv('cah2_y_test.csv',delimiter='\t',names=['Y'])
ylr_pred = pd.read_csv('cah2_ylr_pred.csv',delimiter='\t',names=['Y'])
yrfr_pred = pd.read_csv('cah2_yrfr_pred.csv',delimiter='\t',names=['Y'])
ygb_pred = pd.read_csv('cah2_ygb_pred.csv',delimiter='\t',names=['Y'])
```

```python
# Linear Regression
plt.scatter(y_test,ylr_pred, marker='x')

lims = [
    np.min([plt.xlim(), plt.ylim()]),  # min of both axes
    np.max([plt.xlim(), plt.ylim()]),  # max of both axes
]

plt.plot(lims, lims, 'k-')
plt.xlabel('pChEBML Values  Test')
plt.ylabel('pChEBML Values  Predicted')
plt.title('Linear Regression')
plt.tight_layout()
plt.savefig('lin-reg-y-cah2.png')
```

```python
plt.cla()
# Random Forest
plt.scatter(y_test,yrfr_pred, marker='x')

lims = [
    np.min([plt.xlim(), plt.ylim()]),  # min of both axes
    np.max([plt.xlim(), plt.ylim()]),  # max of both axes
]

plt.plot(lims, lims, 'k-')
plt.xlabel('pChEBML Values  Test')
plt.ylabel('pChEBML Values  Predicted')
plt.title('Random Forest')
plt.tight_layout()
plt.savefig('random-forest-y-cah2.png')
```

```python
plt.cla()
# Gradient Boost
plt.scatter(y_test,ygb_pred, marker='x')

lims = [
    np.min([plt.xlim(), plt.ylim()]),  # min of both axes
    np.max([plt.xlim(), plt.ylim()]),  # max of both axes
]

plt.plot(lims, lims, 'k-')
plt.xlabel('pChEBML Values  Test')
plt.ylabel('pChEBML Values  Predicted')
plt.title('Gradient Boost')
plt.tight_layout()
plt.savefig('gboost-y-cah2.png')
```
